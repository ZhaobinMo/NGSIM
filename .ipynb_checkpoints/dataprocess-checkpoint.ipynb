```python
# get general info --> generate time space grid

# set the cell size
import numpy as np
global l_t; global l_s
global grid_t, grid_s
l_t = 30   # second
l_s = 30  # m
import matplotlib.pyplot as plt
global road_segmentMeter
#road_segmentFeet = [0, 578, 578+698, 578+698+824]    # 0.0, 176.17440000000002, 388.9248, 677.6728128
road_segmentFeet = [0, 420, 420+1230]
road_segmentMeter = [0.3048*x for x in road_segmentFeet]
import pandas as pd
from math import floor

global MIN_t_diff_us101
global MAX_t_diff_us101
global MIN_local_y_us101
global MAX_local_y_us101

global MIN_t_diff_i80 
global MAX_t_diff_i80 
global MIN_local_y_i80 
global MAX_local_y_i80

MIN_t_diff_us101 = 0.0
MAX_t_diff_us101 = 2772.5
MIN_local_y_us101 = 0
MAX_local_y_us101 = 681.3048096

MIN_t_diff_i80 = -0.0
MAX_t_diff_i80 = 5597.9
MIN_local_y_i80 = 0
MAX_local_y_i80 = 545.8270008000002


def getTimeSpaceGrid(df):
    min_local_y = MIN_local_y_us101#min(df['local_y'])
    max_local_y = MAX_local_y_us101#max(df['local_y'])
    min_t_diff = MIN_t_diff_us101#min(df['t_diff'])
    max_t_diff = MAX_t_diff_us101#max(df['t_diff'])

    # genereate grid
    n_s = (max_local_y - min_local_y) / l_s
    n_s = round(n_s)
    n_t = (max_t_diff - min_t_diff) / l_t
    n_t = round(n_t)

    dif_s = (max_local_y - min_local_y) / n_s
    dif_t = (max_t_diff - min_t_diff) / n_t
    grid_t = list(map(lambda x: min_t_diff+dif_t*x,   range(0,n_t+1)))
    grid_s = list(map(lambda x: min_local_y+dif_s*x,   range(0,n_s+1)))
    return grid_t, grid_s
    
def pointOfTimeGrid(x):
    if x < grid_t[0]:
        idx = 0
    elif x >= grid_t[-1]:
        idx = len(grid_t) - 2
    else:
        dis = x - grid_t[0]
        idx = floor(dis/(grid_t[1]-grid_t[0]))
    return (idx, grid_t[idx], grid_t[idx+1])

def pointOfSpaceGrid(x):
    if x < grid_s[0]:
        idx = 0
    elif x >= grid_s[-1]:
        idx = len(grid_s) - 2
    else:
        dis = x - grid_s[0]
        idx = floor(dis/(grid_s[1]-grid_s[0]))
    return (idx, grid_s[idx], grid_s[idx+1])

def pointOfSegment(x):
    L = road_segmentMeter
    if (x>=L[0])&(x<L[1]):
        return 0
    elif (x>=L[1])&(x<=L[2]):
        return 1
    elif (x>=L[2]):
        return 2
grid_t, grid_s = getTimeSpaceGrid(df)
print(grid_t)
print(grid_s)
```

```python
# add noticer in the df

def gridIndexLabel(df):
    df['point_t'] = df['t_diff'].apply(pointOfTimeGrid)
    df[['idx_t','low_bound_t','high_bound_t']] = pd.DataFrame(df['point_t'].values.tolist(), index=df.index)
    df['point_s'] = df['local_y'].apply(pointOfSpaceGrid)
    df[['idx_s','low_bound_s','high_bound_s']] = pd.DataFrame(df['point_s'].values.tolist(), index=df.index)
    #del(df['point_t'], df['point_s'])
    return(df)

IS_TEST = 1
if IS_TEST:
    df = dfSet
    df = gridIndexLabel(df)
    df.head(5)
    mask = (df['low_bound_t'] - df['t_diff'])*(df['high_bound_t'] - df['t_diff']) >0
    dfshow = df.loc[mask,:]
```

```python
# find the valid set for later calculation
# non-grid-level filter is okay
    # extract the continuous vehicle
    
def getValidGdf(df):
    df['idx_tuple'] = list(zip(df.vehicle_id, df.v_class, df.idx_t, df.idx_s))
    whole_set = set(list(df['idx_tuple']))

    # not-contiunose set
    '''There is a trick: the 4th mask must be NOT continuose, because
    the traj. is not continuouse even if there is just 1 uncontinuosue point,
    which is easier to prove that all the traj. is continuous'''

    #mask_is1toNminus1_unit_id = ~df['unit_id'].ne(df['unit_id'].shift(-1)) # location of 1 to n-1 for each 'unit_id' segment
    mask_is1toNminus1_vechiel_id = ~df['vehicle_id'].ne(df['vehicle_id'].shift(-1))
    mask_is1toNminus1_v_class = ~df['v_class'].ne(df['v_class'].shift(-1))
    mask_is1toNminus1_idx_t = ~df['idx_t'].ne(df['idx_t'].shift(-1)) # location of 1 to n-1 for each 'idx_t' segment
    mask_is1toNminus1_idx_s = ~df['idx_s'].ne(df['idx_s'].shift(-1)) # location of 1 to n-1 for each 'idx_s' segment
    mask_notContinuous = (df['t_diff'].shift(-1) - df['t_diff'] > 1) # no gap in the time series
    mask_all = mask_is1toNminus1_vechiel_id&mask_is1toNminus1_v_class&mask_is1toNminus1_idx_t&mask_is1toNminus1_idx_s&mask_notContinuous
    notContinuousSet = set(list(df.loc[mask_all, 'idx_tuple']))


    # grid-level calculation
    grouped = df.groupby(['vehicle_id', 'v_class', 'idx_t','idx_s'])
    gdf = pd.DataFrame(grouped['local_y'].min()); gdf.columns = ['st_local']; 
    gdf['ed_local'] = grouped['local_y'].max()
    gdf['st_tdiff'] = grouped['t_diff'].min()
    gdf['ed_tdiff'] = grouped['t_diff'].max()
    gdf['low_bound_t'] = grouped['low_bound_t'].first()
    gdf['high_bound_t'] = grouped['high_bound_t'].first()
    gdf['low_bound_s'] = grouped['low_bound_s'].first()
    gdf['high_bound_s'] = grouped['high_bound_s'].first()
    temp_gdf = gdf.reset_index()
    gdf['idx_tuple'] = list(zip(temp_gdf['vehicle_id'], temp_gdf['v_class'], temp_gdf['idx_t'], temp_gdf['idx_s']))

    # the traj. that is from the begining to the end
    mask_consistent = (gdf['high_bound_s'] - gdf['ed_local'] <= 10 ) & (gdf['st_local'] - gdf['low_bound_s'] <= 10 )
    consistentSet = set(list(gdf.loc[mask_consistent,'idx_tuple']))

    # the idx for valid for the speed calculation
    validSet = (whole_set - notContinuousSet) and (consistentSet)

    # Then calculathe the speed the each (index1,2,3) element
    mask_valid = gdf['idx_tuple'].isin(validSet)
    valid_gdf = gdf.loc[mask_valid, :]
    valid_gdf['t_travel'] = valid_gdf['ed_tdiff'] - valid_gdf['st_tdiff']
    print('ratio of valid set: {}'.format(len(validSet)/len(whole_set)))
    print('ratio of notContinuousSet: {}'.format(len(notContinuousSet)/len(whole_set)))
    print('ratio of consistentSet: {}'.format(len(consistentSet)/len(whole_set)))

    # get valid_ggdf
    valid_gdf = valid_gdf.reset_index()

    grouped = valid_gdf.groupby(['idx_t','idx_s'])
    valid_ggdf = pd.DataFrame({'num_cars': grouped['vehicle_id'].count()})
    valid_ggdf['sum_t_travel'] = grouped['t_travel'].sum()

    # remove the 0
    mask = valid_ggdf['sum_t_travel'] == 0
    valid_ggdf = valid_ggdf.loc[~mask,:]
    valid_ggdf['ave_t_travel'] = valid_ggdf['sum_t_travel'] / (valid_ggdf['num_cars'])
    valid_ggdf['q'] = valid_ggdf['num_cars'] / (grid_t[1]-grid_t[0])
    valid_ggdf['ave_speed'] = (grid_s[1]-grid_s[0]) / valid_ggdf['ave_t_travel']
    valid_ggdf['density'] = valid_ggdf['q'] / valid_ggdf['ave_speed']

    # get seg id
    grid_s_segId = list(map(pointOfSegment, grid_s))
    color_list = ['red', 'green', 'blue']
    valid_ggdf = valid_ggdf.reset_index()
    valid_ggdf['seg_id'] = valid_ggdf['idx_s'].apply(lambda x: grid_s_segId[x]).astype(int)
    valid_ggdf['seg_color'] = valid_ggdf['seg_id'].apply(lambda x: color_list[x])
    return valid_ggdf
valid_ggdf = getValidGdf(df)
```

# Plot the jamiton

```python
V_MIN = 7.5; ROU_MAX = 1/V_MIN; U_MAX = 20;beta = 8;gamma = 1/2
import math
def equilibrium_g(y):
    b = 1/3
    Lambda = 1/10
    return math.sqrt(1+((y-b)/Lambda)**2)

def equilibrium_Q(rou):
    c = ROU_MAX * U_MAX
    return c*( equilibrium_g(0) + (equilibrium_g(1)-equilibrium_g(0))*rou/ROU_MAX - equilibrium_g(rou/ROU_MAX) )

def hesitation_h(rou):
    return belta*( (rou/ROU_MAX)/(1-rou/ROU_MAX) )**gamma

def U(rou):
    return U_MAX( 1-rou/ROU_MAX )

def U_hat(v):
    return U_MAX*( 1 - V_MIN/v )

def SCC_d1U(rou):
    return -U_MAX/ROU_MAX

def SCC_d1h(rou):
    return belta*gamma*( (rou/ROU_MAX)/(1-rou/ROU_MAX) )**(gamma-1) / (ROU_MAX*(1 - rou/ROU_MAX)**2)
def SCC(rou):
    if SCC_d1h(rou) > -1*SCC_d1U(rou):
        return True
    else:
        return False

def h_hat_d1(v):
    return belta*V_MIN**gamma*(-1*gamma)*(v-V_MIN)**(-1*(gamma + 1))
    
```

```python
class Jamiton(object):
    '''To find the jamiton segment endpoint
    Input:      -v    every v in the (v_min, +inf)
    
    Output:     -m    
                -s
                -rou_M    to get the left jamiton point(rou_M, m+s*rou_M)
                -rou_R    to get the right jamiton point(rou_R, m+s*rou_R)
    '''
    def __init__(self, ROU_MAX=1/7.5, U_MAX=20, beta=8, gamma=1/2):
        self.V_MIN = 1/ROU_MAX
        self.ROU_MAX = ROU_MAX
        self.U_MAX = U_MAX
        self.beta = beta
        self.gamma = gamma
        self.m = 1
        self.s = 1
    
    # U functions
    def dU(self, rou):
        U_MAX = self.U_MAX; ROU_MAX = self.ROU_MAX
        return -U_MAX/ROU_MAX
    
    def U_hat(self, v):
        U_MAX = self.U_MAX; V_MIN = self.V_MIN
        return U_MAX*( 1 - V_MIN/v )
    
    def dU_hat(self, v):
        V_MIN = self.V_MIN
        return V_MIN/(v**2)
    
    # h functions
    def dh(self, rou):
        beta = self.beta; gamma = self.gamma; ROU_MAX = self.ROU_MAX
        #return beta*gamma*( (rou/ROU_MAX)/(1-rou/ROU_MAX) )**(gamma-1) / (ROU_MAX*(1 - rou/ROU_MAX)**2)
        return beta*gamma* (rou/(ROU_MAX-rou))**(gamma-1) * ROU_MAX/(ROU_MAX - rou)**2
        
    def h_hat(self, v):
        beta = self.beta; V_MIN = self.V_MIN; gamma = self.gamma
        return beta*(V_MIN/(v-V_MIN))**gamma
    
    def dh_hat(self, v):
        V_MIN = self.V_MIN; gamma = self.gamma; beta = self.beta
        return beta*V_MIN**gamma*(-1*gamma)*(v-V_MIN)**(-1*(gamma + 1))
    
    # get the m and s first, and then we can construct the other function
    # const_ms: get the m, s for a specific v
    # note whether use it or not it controled by the getVM. Won't use it if v is not sonic ponit
    def const_ms(self,vs):
        m = -1*self.dh_hat(vs)
        s = self.U_hat(vs) - m*vs
        self.m = m
        self.s = s
        #return m, s
    
    # w functions
    def w(self, v):
        m = self.m; s = self.s
        return self.U_hat(v) - (m*v+s)
    
    def dw(self, v):
        m = self.m
        return self.dU_hat(v) - m
    
    # r function
    def r(self, v):
        m = self.m;
        return m*self.h_hat(v) + m**2 * v
    
    # checkSCC whther v is the sonic point or not
    def checkSCC(self, rou):
        if self.dh(rou) > -1*self.dU(rou):
            #print('SCC is true')
            return True
        else:
            #print('SCC is false')
            return False
    
    # checkDw: dw
    def checkDw(self, v):
        if self.dw(v) <= 0:
            #print('check DW is false')
            return False
        if self.dw(v) > 0:
            #print('Check DW is true')
            return True
            
    #  get the vm
    def getVm(self, v):
        U_MAX = self.U_MAX; s = self.s; m = self.m
        return (U_MAX - s)/m - v
    
    # get the vr
    def getVr(self, Vm, Vs):
        thr = 1e-3
        maxIter = 1000; n = 0
        L = V_MIN
        R = Vs    # left and right point fot the dichotomization
        flag = 0; 

        # initialization
        Vr = (L + R )/2
        D = abs(self.r(Vr) - self.r(Vm))

        while D>thr:
            if self.r(Vr) > self.r(Vm):
                L = Vr
                Vr = (L + R)/2
            else:
                R = Vr
                Vr = (L + R)/2
            n = n + 1
            D = abs(self.r(Vr) - self.r(Vm))
            if n > maxIter:
                print('unprecise Vr at Vs = {}, Vr = {}, D = {}, L ={}, R = {}'.format(Vs, Vr, D, L, R))
                break
        return Vr
    
    # The "main" func
    def parameter(self, ROU_MAX=1/7.5, U_MAX=20, beta=8, gamma=1/2):
        self.V_MIN = 1/ROU_MAX
        self.ROU_MAX = ROU_MAX
        self.U_MAX = U_MAX
        self.beta = beta
        self.gamma = gamma
    
    def cal(self, rou):
        v = 1/rou
        # get m and s
        self.const_ms(v)
        m = self.m; s = self.s
        if ((self.checkSCC(rou)) & (self.checkDw(v))) | True:
            # get vm
            vm = self.getVm(v)
            # get vr
            vr = self.getVr(vm,v)
            
            return m, s, vm, vr
        else:
            return False
            
        
```

```python
# Jamiton with new Q function
class Jamiton_PiecewiseU(object):
    '''To find the jamiton segment endpoint
    Input:      -v    every v in the (v_min, +inf)
    
    Output:     -m    
                -s
                -rou_M    to get the left jamiton point(rou_M, m+s*rou_M)
                -rou_R    to get the right jamiton point(rou_R, m+s*rou_R)
    '''
    def __init__(self, ROU_MAX=1/7.5, U_MAX=20, beta=8, gamma=1/2):
        self.V_MIN = 1/ROU_MAX
        self.ROU_MAX = ROU_MAX
        self.U_MAX = U_MAX
        self.beta = beta
        self.gamma = gamma
        self.m = 1
        self.s = 1
    
    # U functions
    def dU(self, rou):
        U_MAX = self.U_MAX; ROU_MAX = self.ROU_MAX
        return -U_MAX/ROU_MAX
    
    def U_hat(self, v):
        U_MAX = self.U_MAX; V_MIN = self.V_MIN
        return U_MAX*( 1 - V_MIN/v )
    
    def dU_hat(self, v):
        V_MIN = self.V_MIN
        return V_MIN/(v**2)
    
    # h functions
    def dh(self, rou):
        beta = self.beta; gamma = self.gamma; ROU_MAX = self.ROU_MAX
        #return beta*gamma*( (rou/ROU_MAX)/(1-rou/ROU_MAX) )**(gamma-1) / (ROU_MAX*(1 - rou/ROU_MAX)**2)
        return beta*gamma* (rou/(ROU_MAX-rou))**(gamma-1) * ROU_MAX/(ROU_MAX - rou)**2
        
    def h_hat(self, v):
        beta = self.beta; V_MIN = self.V_MIN; gamma = self.gamma
        return beta*(V_MIN/(v-V_MIN))**gamma
    
    def dh_hat(self, v):
        V_MIN = self.V_MIN; gamma = self.gamma; beta = self.beta
        return beta*V_MIN**gamma*(-1*gamma)*(v-V_MIN)**(-1*(gamma + 1))
    
    # get the m and s first, and then we can construct the other function
    # const_ms: get the m, s for a specific v
    # note whether use it or not it controled by the getVM. Won't use it if v is not sonic ponit
    def const_ms(self,vs):
        m = -1*self.dh_hat(vs)
        s = self.U_hat(vs) - m*vs
        self.m = m
        self.s = s
        #return m, s
    
    # w functions
    def w(self, v):
        m = self.m; s = self.s
        return self.U_hat(v) - (m*v+s)
    
    def dw(self, v):
        m = self.m
        return self.dU_hat(v) - m
    
    # r function
    def r(self, v):
        m = self.m;
        return m*self.h_hat(v) + m**2 * v
    
    # checkSCC whther v is the sonic point or not
    def checkSCC(self, rou):
        if self.dh(rou) > -1*self.dU(rou):
            #print('SCC is true')
            return True
        else:
            #print('SCC is false')
            return False
    
    # checkDw: dw
    def checkDw(self, v):
        if self.dw(v) <= 0:
            #print('check DW is false')
            return False
        if self.dw(v) > 0:
            #print('Check DW is true')
            return True
            
    #  get the vm
    def getVm(self, v):
        U_MAX = self.U_MAX; s = self.s; m = self.m
        return (U_MAX - s)/m - v
    
    # get the vr
    def getVr(self, Vm, Vs):
        thr = 1e-3
        maxIter = 1000; n = 0
        L = V_MIN
        R = Vs    # left and right point fot the dichotomization
        flag = 0; 

        # initialization
        Vr = (L + R )/2
        D = abs(self.r(Vr) - self.r(Vm))

        while D>thr:
            if self.r(Vr) > self.r(Vm):
                L = Vr
                Vr = (L + R)/2
            else:
                R = Vr
                Vr = (L + R)/2
            n = n + 1
            D = abs(self.r(Vr) - self.r(Vm))
            if n > maxIter:
                print('unprecise Vr at Vs = {}, Vr = {}, D = {}, L ={}, R = {}'.format(Vs, Vr, D, L, R))
                break
        return Vr
    
    # The "main" func
    def parameter(self, ROU_MAX=1/7.5, U_MAX=20, beta=8, gamma=1/2):
        self.V_MIN = 1/ROU_MAX
        self.ROU_MAX = ROU_MAX
        self.U_MAX = U_MAX
        self.beta = beta
        self.gamma = gamma
    
    def cal(self, rou):
        v = 1/rou
        # get m and s
        self.const_ms(v)
        m = self.m; s = self.s
        if ((self.checkSCC(rou)) & (self.checkDw(v))) | True:
            # get vm
            vm = self.getVm(v)
            # get vr
            vr = self.getVr(vm,v)
            
            return m, s, vm, vr
        else:
            return False
            
```
